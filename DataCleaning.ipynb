# Data cleaning

```python
# deleting the columns with categorical data types, as these attributes have been deemed unimportant 
del df['IDATE']
del df['IMONTH']
del df['IDAY']
del df['IYEAR']
del df['PCDMDECN']
del df['EXACTOT1']
del df['EXACTOT2']
```

```python
# checking value types and missing data 
df.info()
```

```python
# checking how many missing values each attribute has 
# notice that only 79/323 attributes have no missing values
df.isnull().sum().sort_values()
```

```python
# select important columns
df_important = df[['_MICHD',
                   '_RFHYPE5',
                   'TOLDHI2',
                   '_CHOLCHK', 
                   '_BMI5', 
                   'SMOKE100', 
                   'CVDSTRK3',
                   'DIABETE3', 
                   '_TOTINDA', 
                   '_FRTLT1',
                   '_VEGLT1', 
                   '_RFDRHV5', 
                   'HLTHPLN1',
                   'MEDCOST', 
                   'GENHLTH',
                   'MENTHLTH',
                   'PHYSHLTH',
                   'DIFFWALK', 
                   'SEX',
                   '_AGEG5YR',
                   'EDUCA',
                   'INCOME2' ]]
df_important.shape
```

```python
# dropping any remaining missing values 
df_important = df_important.dropna()
df_important.shape
```

```python
# refering to the dataset, 2 means did not have so 0 is a better indicator
df_important['_MICHD'] = df_important['_MICHD'].replace({2: 0})
df_important._MICHD.unique()
```

```python
# 1 means no high blood pressure so change it to 0
# 2 means high blood pressure so change it to 1 
# 9 means refused so eliminate
df_important['_RFHYPE5'] = df_important['_RFHYPE5'].replace({1:0, 2:1})
df_important = df_important[df_important._RFHYPE5 != 9]
df_important._RFHYPE5.unique()
```

```python
# 2 means doesn't have it so change to 0
# 7 means don't know so eliminate
# 9 means refused so eliminate
df_important['TOLDHI2'] = df_important['TOLDHI2'].replace({2:0})
df_important = df_important[df_important.TOLDHI2 != 7]
df_important = df_important[df_important.TOLDHI2 != 9]
df_important.TOLDHI2.unique()
```

```python
# Three means cholesterol hasn't been checked in the past 5 years so change to zero
# 2 means XXX so change to 0
# 9 means refused so eliminate
df_important['_CHOLCHK'] = df_important['_CHOLCHK'].replace({3:0,2:0})
df_important = df_important[df_important._CHOLCHK != 9]
df_important._CHOLCHK.unique()
```

```python
# In the dataset BMI is x100, so divide by 100
df_important['_BMI5'] = df_important['_BMI5'].div(100).round(0)
df_important._BMI5.unique()
```

```python
# 2 means don't have so turn to 0
# 7 means don't know so eliminate
# 9 means refused so eliminate
df_important['SMOKE100'] = df_important['SMOKE100'].replace({2:0})
df_important = df_important[df_important.SMOKE100 != 7]
df_important = df_important[df_important.SMOKE100 != 9]
df_important.SMOKE100.unique()
```

```python
# 2 means don't have so turn to 0
# 7 means don't know so eliminate
# 9 means refused so eliminate
df_important['CVDSTRK3'] = df_important['CVDSTRK3'].replace({2:0})
df_important = df_important[df_important.CVDSTRK3 != 7]
df_important = df_important[df_important.CVDSTRK3 != 9]
df_important.CVDSTRK3.unique()
```

```python
# Change so 0 = no long term diabetes, 1 = pre-diabetic, 2 = has diabetes
# 7 means don't know so eliminate
# 9 means refused so eliminate
df_important['DIABETE3'] = df_important['DIABETE3'].replace({2:0, 3:0, 1:2, 4:1})
df_important = df_important[df_important.DIABETE3 != 7]
df_important = df_important[df_important.DIABETE3 != 9]
df_important.DIABETE3.unique()
```

```python
# 2 means no physical activity so change to 0
# 9 means refused so eliminate
df_important['_TOTINDA'] = df_important['_TOTINDA'].replace({2:0})
df_important = df_important[df_important._TOTINDA != 9]
df_important._TOTINDA.unique()
```

```python
# 2 means no fruit so change to 0
# 9 means refused so eliminate
df_important['_FRTLT1'] = df_important['_FRTLT1'].replace({2:0})
df_important = df_important[df_important._FRTLT1 != 9]
df_important._FRTLT1.unique()
```

```python
# 2 means no vegetables so change to 0
# 9 means refused so eliminate
df_important['_VEGLT1'] = df_important['_VEGLT1'].replace({2:0})
df_important = df_important[df_important._VEGLT1 != 9]
df_important._VEGLT1.unique()
```

```python
# 1 means no heavy drinking so change to 0
# 2 means heavy drinking so change to 1
# 9 means refused so eliminate
df_important['_RFDRHV5'] = df_important['_RFDRHV5'].replace({1:0, 2:1})
df_important = df_important[df_important._RFDRHV5 != 9]
df_important._RFDRHV5.unique()
```

```python
# 2 means no health care access so change to 0
# 7 means don't know so eliminate
# 9 means refused so eliminate
df_important['HLTHPLN1'] = df_important['HLTHPLN1'].replace({2:0})
df_important = df_important[df_important.HLTHPLN1 != 7]
df_important = df_important[df_important.HLTHPLN1 != 9]
df_important.HLTHPLN1.unique()
```

```python
# 2 means no so change to 0
# 7 means don't know so eliminate
# 9 means refused so eliminate
df_important['MEDCOST'] = df_important['MEDCOST'].replace({2:0})
df_important = df_important[df_important.MEDCOST != 7]
df_important = df_important[df_important.MEDCOST != 9]
df_important.MEDCOST.unique()
```

```python
# 7 means don't know so eliminate
# 9 means refused so eliminate
df_important = df_important[df_important.GENHLTH != 7]
df_important = df_important[df_important.GENHLTH != 9]
df_important.GENHLTH.unique()
```

```python
# 88 means no bad mental health days so change to 0 
# 77 means don't know so eliminate
# 99 means refused so eliminate
df_important['MENTHLTH'] = df_important['MENTHLTH'].replace({88:0})
df_important = df_important[df_important.MENTHLTH != 77]
df_important = df_important[df_important.MENTHLTH != 99]
df_important.MENTHLTH.unique()
```

```python
# 88 means no bad mental health days so change to 0 
# 77 means don't know so eliminate
# 99 means refused so eliminate
df_important['PHYSHLTH'] = df_important['PHYSHLTH'].replace({88:0})
df_important = df_important[df_important.PHYSHLTH != 77]
df_important = df_important[df_important.PHYSHLTH != 99]
df_important.PHYSHLTH.unique()
```

```python
# 2 means no so change to 0
# 7 means don't know so eliminate
# 9 means refused so eliminate
df_important['DIFFWALK'] = df_important['DIFFWALK'].replace({2:0})
df_important = df_important[df_important.DIFFWALK != 7]
df_important = df_important[df_important.DIFFWALK != 9]
df_important.DIFFWALK.unique()
```
